### 1. 理解

- 简述 Python 中的几种数据类型。
Python 中有一下数据类型：整型、浮点型、布尔型、字符串、列表、元组、字典、集合。
整型：整型是指整数，如 1、2、3 等。
浮点型：浮点型是指带小数点的数，如 1.2、3.4 等。
布尔型：布尔型是指 True 和 False 两种值。
字符串：字符串是指由字符组成的序列，如 "Hello World!"。
列表：列表是指由一系列元素组成的序列，如 [1, 2, 3]。
元组：元组是指由一系列元素组成的序列，但是元组的元素不可变，如 (1, 2, 3)。
字典：字典是指由键值对组成的序列，如 {'a': 1, 'b': 2, 'c': 3}。
集合：集合是指由一系列元素组成的序列，但是集合的元素不可重复，如 {1, 2, 3}。

- 数据类型的可变与不可变分别有哪些？
可变类型：列表、字典、集合。
不可变类型：整型、浮点型、布尔型、字符串、元组。

- 元组，列表，字典有没有长度的限制?
元组、列表、字典的长度没有限制。

- 集合有那些特性？
集合的元素不可重复，集合的元素是无序的。

- 分别解释"=","==","+="的含义
"="：赋值运算符，将右边的值赋给左边的变量。
"=="：比较运算符，判断左右两边的值是否相等。
"+="：加等于运算符，将左右两边的值相加后赋给左边的变量。

- 解释 'and','or','not'
"and"：逻辑与运算符，左右两边的值都为 True 时，结果为 True，否则为 False。
"or"：逻辑或运算符，左右两边的值都为 False 时，结果为 False，否则为 True。
"not"：逻辑非运算符，对右边的值取反。

- copy(),deepcopy()的区别是什么？
在Python中，copy()和deepcopy()这两个函数分别提供浅复制和深复制的功能，用于复制复合数据类型（如列表或字典）中的元素。
浅复制（Shallow Copy）： 使用copy()函数进行浅复制时，它只复制被复制对象的最外层，内层的元素只复制其引用。如果被复制的最外层元素是不可变对象（如数字，字符串或元组），那么改变被复制对象的值，复制后的对象不受影响；但如果其最外层元素是可变对象（如列表或字典），**当你改变其内层元素的值时，复制后的对象也会受到影响**。
深复制（Deep Copy）: 使用deepcopy()函数时，它会复制被复制对象的所有层次，即对最外层和所有内层元素都进行复制，不仅仅是引用。这样的话，**不论被复制对象的任何层次元素的改动，都不会影响到深复制后的对象**。

```python
# 求 1-100 之间能被 7 整除，但不能同时被 5 整除的所有整数。
for i in range(1, 101):
    if i % 7 == 0 and i % 5 != 0:
        print(i, end=' ')
```

```python
# 输出 1000 以内所有的"⽔仙花数"，所谓"⽔仙花数"是指⼀个三位数，其各个位数字⽴⽅和等于该数本身。
# 例如：153 是⼀个"⽔仙花数"，因为 153=1 的三次⽅＋5 的三次⽅＋3 的三次⽅
for i in range(100, 1000):
    a = i // 100
    b = i // 10 % 10
    c = i % 10
    if i == a ** 3 + b ** 3 + c ** 3:
        print(i, end=' ')
```

```python
# 利用 if 语句写出猜大小的游戏。
# 预设⼀个 0-9 的随机整数，让⽤户输⼊所猜的数。如果⼤于预设的数，提示“bigger”；如果⼩于预设的数，显示“smaller”。如此循环，直⾄猜中该数，显示“right！”，并统计第几次猜中。
import random

num = random.randint(0, 9)
count = 0
while True:
    guess = int(input("请输入一个0-9的整数："))
    count += 1
    if guess > num:
        print("bigger")
    elif guess < num:
        print("smaller")
    else:
        print("right!")
        break
print("你猜了%d次" % count)
```

用Pandas实现以下统计算法，并且要求必须从公式层面掌握：
- 均值（Mean）：各数值加起来除以数值的总个数。 公式：$Mean=\frac{1}{n}\sum_{i=1}^{n}{x_i}$
- 中位数（Median）：数据按大小排序后，位于中间的数值。如果数据个数为偶数，则中位数为中间两个数的平均值。
- 众数（Mode）与异众比率：数据集中出现频率最高的数值称为众数。一个数据集可以有一个众数、多个众数或者没有众数（所有数据点均不同）。异众比率是描述数据集中最常见的值的频率，计算公式为最常见的值的频率除以所有频率的总和。
- 极差（Range）：数据的最大值和最小值之差。 公式：$Range = Max(x) - Min(x)$
- 方差（Variance）和标准差（Standard Deviation） 方差：各个数据与平均数之差的平方的平均数。 标准差：方差的平方根。 公式：$Variance = \frac{1}{n}\sum_{i=1}^{n}{(x_i - Mean)^2}$， $Standard\ Deviation=\sqrt{Variance}$
- 协方差（Covariance）：两个变量共同变化的程度。如果两个变量共同增加或减少（例如身高和体重），则协方差是正数；如果一个变量增大而另一个减小，则协方差为负。协方差为0表示两个变量之间没有线性关系。 公式： $ Cov(X,Y) = \frac{1}{n}\sum_{i=1}^{n}{(x_i - Mean_X) (y_i - Mean_Y)} $
- 一维离散数据概率分布直方图：描述单变量的直方图。
- 正态分布的拟合：利用正态分布（也称高斯分布）的概率密度函数来拟合数据。正态分布的概率密度函数为： $f(x) = \frac{1}{\sqrt{2 \pi \sigma^2}} e^{-\frac{(x - \mu)^2}{2 \sigma^2}}$
其中，$\mu$代表均值，$\sigma$代表标准差。

### 根据 US_Crime_Rates_1960_2014.csv 完成任务

使用 US_Crime_Rates_1960_2014.csv，根据 1991-2014 年的数据制作一条回归曲线，预测 2019 年的各个犯罪条目的犯罪率。

```python
data = pd.read_csv('./data/US_Crime_Rates_1960_2014.csv', index_col=0)
data.head()
```

```python
# 计算各个犯罪条目的犯罪率
crime_cols = ['Violent', 'Property', 'Murder', 'Forcible_Rape', 'Robbery',
              'Aggravated_assault', 'Burglary', 'Larceny_Theft', 'Vehicle_Theft']

for col in crime_cols:
    data[f'{col}_rate'] = data[col] / data['Population']
```

```python
data.head()
```

```python
rate_cols = [f'{x}_rate' for x in crime_cols]
rate_cols
```

```python
# 构建单变量时序特征，时间窗口为5，即只使用五年的数据，预测五年后的数据
def time_seqs(df):
    df_time_seq = pd.DataFrame()
    # 预测五年之后的犯罪率
    for i in range(5):
        df_time_seq[f'rate_lag_{i}'] = df.shift(-i)

    df_test = df_time_seq.loc[2010]
    df_time_seq['rate'] = df.shift(-9)

    df_train = df_time_seq.dropna()
    return df_train, df_test
```

```python
def crime_rate_forecast(df, rate_cols):
    for col in rate_cols:
        train, test = time_seqs(df[col])

        feature_cols = [f'rate_lag_{x}' for x in range(5)]
        X_train = train[feature_cols]
        y_train = train['rate']
        X_test = test.to_frame().T

        lr = LinearRegression()
        lr.fit(X_train, y_train)
        pred_train = lr.predict(X_train)
        pred_test = lr.predict(X_test)

        data[col].plot(label='true')
        plt.plot(train.index + 9, pred_train, label='train prediction')
        plt.plot(2019, pred_test, 'rx', label='test prediction')
        plt.title(col)
        plt.legend()
        plt.show()
```

```python
crime_rate_forecast(data, rate_cols)
```
